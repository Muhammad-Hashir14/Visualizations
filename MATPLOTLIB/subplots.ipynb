```python
fig, mulax = plt.subplots(2,2,figsize = (10,6))
fig.subplots_adjust(hspace=0.3)
x = np.arange(1,6)
y = np.random.randint(30,size = 5)
z = np.random.randint(30,size = 5)
m = np.random.randint(30,size = 5)
n = np.random.randint(30,size = 5)
mulax[0,0].plot(x,y,'green',label = 'Green line')
mulax[1,0].plot(x,m,'red', label = 'Red line')
mulax[0,1].plot(x,z,'purple',label = 'purple line')
mulax[1,1].plot(x,n,'orange',label = 'orange line')
for i in range(2):
    for j in range(2):
        mulax[i,j].set_xlabel('x value')
        mulax[i,j].set_ylabel('y value')
        mulax[i,j].legend()
        
```

```python
ar = {
    'a': np.arange(1,31),
    'b': np.random.randint(50,size = 30),
    'c': np.random.randint(50,size = 30),
    'd': np.random.randint(50,size = 30)
}
ar['b'] = ar['b'] * np.arange(30)
fig, ax = plt.subplots(figsize = (5,2.7),layout = 'constrained')
ax.scatter('a','b', c= 'c',s = 'd',data = ar)
```

```python
fig, ax = plt.subplots(figsize = (5,2.7))
x = np.arange(10)
ax.plot(x,x,label = 'linear')
ax.plot(x, x**2,'green',label = 'Squared')
ax.plot(x, x**3,'brown',label = 'Cubic')
ax.set_xlabel('x values')
ax.set_ylabel('y values')
ax.legend()
```

```python
fig, ax = plt.subplots(figsize = (5,2.7))
x = np.arange(10)
y = np.random.randint(0,8,size = 10)
z = np.random.randint(9,16,size = 10)
m = np.random.randint(17,24,size = 10)
n = np.random.randint(24,32,size = 10)
ax.plot(x,y,'o',label = 'circles')
ax.plot(x,z,'v',label = 'triangles')
ax.plot(x,m,'s',label = 'squares')
ax.plot(x,n,'d',label = 'diamonds')
ax.legend()
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7))
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2 * np.pi * t)
line, = ax.plot(t, s, lw=2)
ax.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
arrowprops=dict(facecolor='green', shrink=0.08))

ax.set_ylim(-2, 2);
print(np.pi)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(5, 2.7),layout = 'constrained')
x = np.arange(50)
xdata = np.arange(len(x))  # make an ordinal for this
data = 10**x
axs[0].plot(xdata, data)

axs[1].set_yscale('log')
axs[1].plot(xdata, data);
```

```python
fig, ax = plt.subplots(2,1)
fig.subplots_adjust(hspace = 0.5)
x = np.arange(30)
y = np.random.randint(50,size = 30)
z = np.random.randint(50,size = 30)
ax[0].plot(x,y)
ax[1].plot(x,z)
for i in range(2):
    ax[i].set_xlabel('x values')
    ax[i].set_ylabel('y values')
ax[0].set_title('Automatic Ticks')
ax[1].set_title('Manual Ticks')
ax[1].set_xticks(np.arange(0, 30, 6), ['zero', 6, 'twelve', 18,'twenty four'])
ax[1].plot([0,30],[20,20])
```
